```python
# stopwords extracting
def stopword():
    stop_words = set()
    #iterating files in the directory and finding stopword file
    for file in os.listdir():
      if file == 'all_stop_words.txt' :
        with open(file, 'r', encoding=  'ISO-8859-1') as f:
          stop_words.update(f.read().splitlines())
    return stop_words
stop_words = stopword()
print(stop_words)
# convert to lowercase
stop_words = [word.lower() for word in stop_words]
print(stop_words)
```

```python
# loading the positive and negative word lists and computing the variables
def positive_negative():
  pos_words = set()
  neg_words = set()

  for file in os.listdir():
    if file == 'positive-words.txt' :
      with open(file,'r', encoding ='ascii') as f:
        pos_words.update(f.read().splitlines())
    elif file == 'negative-words.txt' :
      with open(file,'r', encoding = 'ISO-8859-1') as f:
        neg_words.update(f.read().splitlines())
  return pos_words, neg_words
pos_words, neg_words = positive_negative()
print("Positive Words:", pos_words)
print("Negative Words:", neg_words)
# all words in both lists are in lowercase already
```

```python
# function to process the text
# file to lower case
def lower_case_text(file_text):
  file_text = re.sub(r'[^\w\s]', '', file_text.lower()) # remove puctuations
  return file_text
```

The code below first remove the stopwords from the text. Usually, the first step is always LEMMATIZATION which include changing words to their base values or STEMMING which is also quite the same. But doing so would make a difference in calculating the variables taking each form of word as different.
Difference between LEMMATIZATION and STEMMING :
Lemmatization considers the meaning of the word after reduction but Stemming does not.
ex. caring lemmatize to care but stem to car.

```python
# compute variables

def final_var( file_name, file_text, pos_words, neg_words,stop_words):
  clean_text = lower_case_text(file_text)
  text_tokens = word_tokenize(clean_text) # tokenizing the word


  # starting to preprocess the text
  # 1. REMOVE STOPWORDS FROM STOPWORDS FILES
  final_tokens =[word for word in text_tokens if word not in stop_words]
  total_words =len(final_tokens)
  # variables
  positive_score = sum(1 for word in final_tokens if word in pos_words)
  negative_score = sum(1 for word in final_tokens if word in neg_words)
  polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)
  subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001) if total_words > 0 else 0
  #
  s = sent_tokenize(clean_text)
  avg_sentence_length = total_words / len(s) if s else 0
  complex_word_count = sum(1 for word in final_tokens if len(re.findall(r'[aeiou]', word)) > 2)
  word_count = total_words
  syllable_count = sum(sum(1 for char in word if char in 'aeiou') for word in final_tokens)
  personal_pronouns = sum(1 for word in final_tokens if word in ['i', 'we', 'my', 'ours', 'us'])
  avg_word_length = sum(len(word) for word in final_tokens) / word_count if word_count else 0

  #
  percentage_complex_words = complex_word_count / word_count if word_count else 0
  fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

  return {
        "url_id" : file_name,
        "Positive Score": positive_score,
        "Negative Score": negative_score,
        "Polarity Score": polarity_score,
        "Subjectivity Score": subjectivity_score,
        "Avg Sentence Length": avg_sentence_length,
        "Avg Number of Words Per Sentence": avg_sentence_length,
        "Complex Word Count": complex_word_count,
        "Word Count": word_count,
        "Syllables Per Word": syllable_count / word_count if word_count else 0,
        "Personal Pronouns": personal_pronouns,
        "Avg Word Length": avg_word_length,
        "Percentage of Complex Words": percentage_complex_words,
        "Fog Index": fog_index,
    }
```

```python
# Last thing is to final move to the zipfile :
nltk.download('punkt_tab')

def process_the_zipfile(zip_name,output_file ='output_file.xlsx'):
  with zipfile.ZipFile(zip_name,'r') as zip_f :
    zip_f.extractall("extracted_texts")
  results=[]
  for file in os.listdir("extracted_texts"):
        file_path = os.path.join("extracted_texts", file)
        print(file_path)
        with open(os.path.join("extracted_texts", file), 'r', encoding='MacRoman') as file:
          file_text =file.read()
          file_name = os.path.basename(file_path)
          analyse = final_var(file_name, file_text, pos_words, neg_words,stop_words)
          results.append(analyse)
          print(f"Processed file: {file_name}")
  df = pd.DataFrame(results)
  df.to_excel(output_file, index= False)
  print(f"Results saved to {output_file}")

zip_name='extracted_texts.zip'
process_the_zipfile(zip_name)


```
